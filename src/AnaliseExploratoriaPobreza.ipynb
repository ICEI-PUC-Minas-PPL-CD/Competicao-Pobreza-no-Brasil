### Matrix de Correlação
- O coeficiente de correlação quantifica o grau e a direção da relação linear entre duas variáveis. **Ele varia de -1 a 1, onde -1 indica uma correlação negativa perfeita, 1 indica uma correlação positiva perfeita e 0 indica nenhuma correlação linear.**
- **Identificar Relacionamentos:** Ela ajuda a identificar quais variáveis têm correlações fortes e quais têm correlações fracas. Isso é útil para entender quais variáveis podem estar relacionadas e, portanto, podem afetar umas às outras.

## Interpolando os Outros Atributos para Análise // Depois de Testar Interpolações

```python
from scipy.interpolate import interp1d
import pandas as pd

# Suponha que você já tenha definido df_BTCO_normal e df_ipea_normal
# e a lista de colunas a serem interpoladas
colunas_a_interpolar = ['Total doadores triados', 'Total triado desq.',
                        'Total escleras desqualificadas pós preservação',
                        'Total de córneas desqualificadas', 'Total córneas preservadas']

# Obtenha o tamanho das duas bases de dados
tamanho_maior = len(df_ipea_normal)
tamanho_menor = len(df_BTCO_normal)

# Crie uma função de interpolação para cada coluna
interpolacoes = {}
for coluna in colunas_a_interpolar:
    interpolacao = interp1d(range(tamanho_menor), df_BTCO_normal[coluna], kind='quadratic', fill_value='extrapolate')
    interpolacoes[coluna] = interpolacao

# Gere os novos índices para os dados interpolados
novos_indices = range(0, tamanho_maior)

# Interpole os valores para cada coluna
valores_interpolados = {}
for coluna, interpolacao in interpolacoes.items():
    valores_interpolados[coluna] = interpolacao(novos_indices)

# Crie uma cópia vazia do DataFrame com o mesmo tamanho de df_ipea_normal
BTCO_interpolada = pd.DataFrame(columns=df_ipea_normal.columns, index=range(tamanho_maior))

# Preencha as colunas desejadas com os valores interpolados
for coluna in colunas_a_interpolar:
    BTCO_interpolada[coluna] = valores_interpolados[coluna]
```
